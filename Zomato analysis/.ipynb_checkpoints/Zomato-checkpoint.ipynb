## ZOMATO Data Analaysis Project

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe=pd.read_csv("Zomato data.csv")
```

```python
dataframe.head()
```

## Data cleaning

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0]
    return float(value)
```

```python
dataframe['rate']=dataframe['rate'].apply(handleRate)
dataframe.head()
```

```python
dataframe.info()
```

## Type of restaurants getting are more 

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of restaurant", c='blue', size=15)
```

## See which type of restaurants receive the highest votes.

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()                    # group_by is applied for summing votes
grouped_data
```

```python
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="green", marker="o")                             #when we have both x and y axis data
plt.xlabel("Type of restaurant", c="blue", size=20)
plt.ylabel("Votes", c="blue", size=20)
```

## Majority of rating received? ( Rating Distribution)

```python
plt.hist(dataframe['rate'],bins=5)                      #make intervals of entries distributed under number of bins
plt.xlabel(' Ratings', c="blue", size=17)
plt.ylabel('Counts', c="blue", size=17)
```

## Average spending on each order by couple?

```python
sns.countplot(x=dataframe['approx_cost(for two people)'])               # make bars of individual entries
plt.xlabel('Amount spent', c="blue", size=17)
plt.ylabel('Counts', c="blue", size=17)
```

## Either offline/online mode received higher rating

```python
plt.figure(figsize=(6,6))                                       # To resize any plot
sns.boxplot(x='online_order', y='rate', data=dataframe)         # or write directly (x=dataframe['online_order'], y=dataframe['rate'])
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)    #counts the occurrences for each cell
# can be summed /averaged in cell             #fill_value where no counts

sns.heatmap(pivot_table, annot=True, cmap="coolwarm", fmt='d')         # Formats the numbers as integers.
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```
